# simple introduction to the lineal regression 

```python
import matplotlib.pyplot as plt
import matplotlib.lines as lines
import numpy as np
plt.style.use("seaborn-v0_8")

def calcula_ganancia(x):
    y = 0.8 * x + 5
    return y

print(calcula_ganancia(10))
```

<div style="font-size:25px">
$$\hat{y} = f(\vec{x}) $$ 
<br>
$$\hat{y} = \vec{\omega} \cdot \vec{x} + b $$ 
<br>
$$ \hat{y} = \omega_1 x_1 + \omega_2 x_2 + \cdots + \omega_n x_n + b$$ 
<br>

```python
# X is for the invested money and Y is for the gainings
x = np.array([0, 6.5, 13.2, 18.1, 24.9, 28.2, 36.1])
y = np.array([5, 10, 17, 20, 25, 30, 35])
```

```python
plt.figure(figsize=(7,7))
plt.scatter(x, y, s=60)
plt.ylim(bottom=0)
plt.title("Values")
plt.xlabel("Invested Money")
plt.ylabel("Income")
plt.grid()
plt.show()
```

```python
x2 = np.arange(0, 40)
y2 = 0.8 * x2
print(y2)
x2
```

```python
plt.figure(figsize=(7,7))
plt.plot(x2, y2, color="red", label="Our solution")
plt.scatter(x, y, s=60)
plt.ylim(bottom=0)
plt.title("Values")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.grid()
plt.show()
```

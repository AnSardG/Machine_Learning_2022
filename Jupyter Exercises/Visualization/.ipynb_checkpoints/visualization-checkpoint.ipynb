### World Development Indicators


The World Development Indicators from the World Bank have more than a thousand annual indicators of economic development of thousands of contruies around the world.

Here's a list of the available indicators next to a list of the available countries.

Source: https://www.kaggle.com/datasets/kaggle/world-development-indicators


```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt

plt.style.use("seaborn-v0_8")
```

```python
data = pd.read_csv('wdi/Indicators.csv')
print(data.shape)
data.head()
```

```python
filter = data['CountryName']=='Spain'
print(data[filter].shape)
data[filter].head()
```

### How many countries are there?

```python
countries = data['CountryName'].unique().tolist()
print("Countries => ",len(countries))
```

### How many countrycode are there?

```python
# Then if the countries and codes match, there are no leaks in the dataframe.
countryCodes = data['CountryCode'].unique().tolist()
print("Country codes => ",len(countryCodes))
print(countryCodes)
```

### And how many indicators are?

```python
indicators = data['IndicatorName'].unique().tolist()
print("Indicators => ",len(indicators))
```

### How many years are being registered?

```python
years = data['Year'].unique().tolist()
print("Years => ",len(years))
```

### What's the temporal range?

```python
print("From",min(years),"to",max(years))
```

# Matplotlib: Basic Graphics

```python
# Name of the indicator => CO2 emissions (metric tons per capita)
hist_indicator = 'CO2 emissions \(metric'
hist_country = 'ESP'

mask1 = data['IndicatorName'].str.contains(hist_indicator)
mask2 = data['CountryCode'].str.contains(hist_country)

# Applying both masks.
total_co2 = data[mask1 & mask2]
total_co2.tail()
```

### Analyzing how the emissions changed over time

```python
# Obtaining years
years = total_co2['Year']
# Obtaining the values of each year
co2 = total_co2['Value']

# Draw
plt.bar(years,co2)
plt.show()
```

```python
# Changing the graphic to linear
plt.plot(total_co2['Year'].values, total_co2['Value'])

#Labeling the axes
plt.xlabel('Year')

# In Y we can put the name of the indicator diectly.
plt.ylabel("CO2 emissions (Netric Tons Per Capita)")

#Labeling the figure
plt.title('CO2 emissions in Spain')

#This way we can retouch the scales
plt.axis([1959, 2011, 0,9])

plt.show()
```

### Using histograms to explore the values distribution

```python
hist_data = total_co2['Value']
print(type(hist_data))histogram
hist_data
```

```python
plt.hist(hist_data, 10, facecolor='green')

plt.xlabel(total_co2['IndicatorName'].iloc[0])
plt.ylabel('# of Years')
plt.title('Histogram')

plt.grid(True)

plt.show()
```

We can see that Spain generates between 5 and 6 metric tons of CO2 per capita most years.

## But... How's Spain compared to the rest of the countries?

```python
# Selecting all countries in the year 2011
hist_indicator = 'CO2 emissions \(metric'
hist_year = 2011

mask1 = data['IndicatorName'].str.contains(hist_indicator)
mask2 = data['Year'].isin([hist_year])

# Applying the masks
co2_2011 = data[mask1 & mask2]
co2_2011.head()
```

```python
print(len(co2_2011))
```

```python
# Let's plot a histogram of the emissions per capita by country
# subplots returns a touple with the figure, axis attributes.
fig, ax = plt.subplots()

ax.annotate("SPAIN",
            xy=(20,9), xycoords='data',
            xytext=(23,30), textcoords='data',
            arrowprops=dict(arrowstyle="->",
                           connectionstyle="arc3"),
            )

plt.hist(co2_2011['Value'], 10, facecolor='green')

plt.xlabel(total_co2['IndicatorName'].iloc[0])
plt.ylabel('# of Countries')
plt.title('Histogram of CO2 Emissions per Capita')

'''
We can adjust momentarily the scale to see the 8,9 aprox that has ESP
and, this way, aim the arrow correctly
'''
#plt.axis([10, 22, 0, 14])
plt.grid(True)

plt.show()
```

```python
# Selecting the GDP or PIB of Spain
hist_indicator = 'GDP per capita \(constant 2005'
hist_country = 'ESP'

mask1 = data['IndicatorName'].str.contains(hist_indicator)
mask2 = data['CountryCode'].str.contains(hist_country)

#Applying the masks
gdp_stage = data[mask1 & mask2]
```

```python
gdp_stage.head(2)
```

```python
# Let's draw a linear graphic
plt.plot(gdp_stage['Year'].values, gdp_stage['Value'].values)

#Label the axes
plt.xlabel('Year')
plt.ylabel(gdp_stage['IndicatorName'].iloc[0])

#Label the figure
plt.title('PIB Per Capita Spain')

# to make it more honest, start the Y axis at 0
plt.axis([1959, 2015, 0, 30000])

plt.show()
```

## Scatter plot to compare PIB vs CO2
For this Scatter plot we're going to assure that PIB and CO2 are being shown on the same temporal windows

```python
print("PIB Min Year =>",gdp_stage['Year'].min(), "Max =>",gdp_stage['Year'].max())
print("CO2 Min Year =>",total_co2['Year'].min(), "Max =>", total_co2['Year'].max())
```

#### Careful... we have 3 additional years between PIB and CO2

```python
gdp_stage_trunc = gdp_stage[gdp_stage['Year']<2012]
print(len(gdp_stage_trunc))
print(len(total_co2))
```

```python
fig, axis = plt.subplots()
# Grid lines, Xticks, Xlabel, Ylabel

axis.yaxis.grid(True)
axis.set_title('CO2 Emissions vs. GDP (per capita)', fontsize=10)
axis.set_xlabel(gdp_stage_trunc['IndicatorName'].iloc[0], fontsize=10)
axis.set_ylabel(total_co2['IndicatorName'].iloc[0], fontsize=10)

X = gdp_stage_trunc['Value']
Y = total_co2['Value']

axis.scatter(X, Y)
plt.show()
```

```python
# Array of value correlations
np.corrcoef(gdp_stage_trunc['Value'], total_co2['Value'])
```
